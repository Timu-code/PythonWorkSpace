```python
import math

# 표준 편차 계산 함수
def std(data):
    n = len(data)
    
    if n < 2:
        return 0
    
    mean = sum(data) / n
    variance = sum((x - mean) ** 2 for x in data) / n
    
    return math.sqrt(variance)

# 유클리드 거리 계산 함수
def EuclidDist(a, b):    
    return round(math.sqrt(sum((a[i] - b[i]) ** 2 for i in range(len(a)))), 3)

# 맨해튼 거리 계산 함수
def ManhattanDist(a, b):
    return round(sum(abs(a[i] - b[i]) for i in range(len(a))), 3)

# 코사인 유사도 계산 함수
def cosine_sim(a, b):
    norm_a = math.sqrt(sum(a[i] ** 2 for i in range(len(a))))
    norm_b = math.sqrt(sum(b[i] ** 2 for i in range(len(b))))
    
    dot_product = sum(a[i] * b[i] for i in range(len(a)))
    
    if norm_a == 0 or norm_b == 0:
        return 0
    
    return round(dot_product / (norm_a * norm_b), 3)

# 피어슨 상관계수 계산 함수
def pearson_corr(a, b):
	n = len(a)
 
	mean_a = sum(a) / n
	mean_b = sum(b) / n
 
	sum_a = sum((x - mean_a) ** 2 for x in a)
	sum_b = sum((x - mean_b) ** 2 for x in b)
	sum_ab = sum((a[i] - mean_a) * (b[i] - mean_b) for i in range(n))
 
	return round(sum_ab / math.sqrt(sum_a * sum_b), 3)
```

```python
# Lines, Line types, Rectangles, Colors, Mondrian (0, 1)
data = [
    [6, 1, 10, 4, 0], # No.1
    [4, 2, 8, 5, 0],  # No.2
    [5, 2, 7, 4, 1],  # No.3
    [5, 1, 8, 4, 1],  # No.4
    [5, 1, 10, 5, 0], # No.5
    [6, 1, 8, 6, 1],  # No.6
    [7, 1, 14, 5, 0]  # No.7
]

new_data = [7, 2, 9, 4]

features = [row[:-1] for row in data]
labels = [row[-1] for row in data]

print(*features)
print(labels)
```

```python
means = []
stds = []

# 각 feature의 평균과 표준 편차 계산
for i in range(len(features[0])):
    column = [row[i] for row in features]
    means.append(sum(column) / len(column))
    stds.append(std(column))
    
print(f'Means: {means}')
print(f'Stds: {stds}')
```

```python
normalized_data = []

# 데이터의 각 값을 평균과 표준 편차로 정규화
for row in features:
    normalized_row = [round((row[i] - means[i]) / stds[i], 3) for i in range(len(row))]
    normalized_data.append(normalized_row)
    
print(*normalized_data, sep='\n')
```

```python
# 새로운 데이터 정규화
normalized_new_data = [round((new_data[i] - means[i]) / stds[i], 3) for i in range(len(new_data))]

print(normalized_new_data)
```

```python
distances = []

# 유클리드 거리 계산
for i in range(len(normalized_data)):
    distance = EuclidDist(normalized_new_data, normalized_data[i])
    distances.append((distance, labels[i]))
    
print(*distances)
```

```python
# 거리 기준 (오름차순) 정렬
distances.sort(key=lambda x: x[0])

print(*distances)
```

```python
k = 5

nearest_neighbors = distances[:k]

# 각 클래스의 수 계산
class_count = {}
classes = set(label for label in labels)

for v in classes:
    current_sum = 0
    
    for _, label in nearest_neighbors:
        
        if v == label:
            delta = 1
        else:
            delta = 0
            
        current_sum += delta
        
    class_count[v] = current_sum
        
# 가장 많은 클래스를 선택
predicted_class = max(class_count, key=class_count.get)
print(f'{new_data}에 대한 분류 결과는 {predicted_class} ({"Yes" if predicted_class == 1 else "No"})')
```
